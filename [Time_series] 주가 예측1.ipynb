```python
# 주식 가격의 예측 인공지능 모델
# LSTM(Long_Short_Term_Memory): 인공지능 자체가 기억(메모리)를 가지고 있다.
# 삼성전자와 이더리움

# 실제 주식 머신러닝 프로그램 설계하는 사람으로서 말씀드리면, 
# 영상만보면 너무 쉽게 생각하실수 있을것 같네요. 
# 실제로는 웹크롤링 기법 및 자동화, 데이터정제 및 분석 , 
# 딥러닝 모델링 등에관한 이론을 모두 알아야합니다. 
# 프로그래밍은 기본이고 자료구조와 알고리즘과 같은 컴퓨터사이언스에 대한 바탕이 있으셔야 
# 효율적인 프로그램을 설계할 수 있습니다. 
# 또한 데이터 사이언스와 베이지안통계쪽의 개념이 있으셔야 제대로 데이터 분석을 할수있고 
# 이러한 데이터 분석과정은 다소 주관이 들어가므로 반드시 도메인에대한 수준높은 이해가 필요합니다.
# 이과정이 있어야만 딥러닝 모델의 결과가 실제로 잘 나오게됩니다.
# 머신러닝 모델 또한 여러가지가있고 lstm만 보더라도 옵션이 굉장히많은데 
# 그러한 것들을 유연하게 조작하기위해서는 기본적으로 대학미적분학과 선형대수를 공부하셔야합니다.
```

```python
import FinanceDataReader as fdr
samsung = fdr.DataReader('005930','2020-03-01') # 삼성전자
e = fdr.DataReader('ETH/USD','2020-03-01') # 이더리움/달러 가격
```

# 1. Compute Mid Price

```python
# 최고가와 최저가의 중간가격 mid_prices
high_prices = samsung['High'].values
Low_prices = samsung['Low'].values
mid_prices = (high_prices + Low_prices)/2
```

# 2. Create Windows


```python
# 최근 n일 데이터를 보고 내일(n+1)을 예측한다.
# 주식 종목에 따라 바꿔줄 수 있다.
seq_len = 50 # 50일 지정
sequence_length = seq_len + 1 

result = []
for index in range(len(mid_prices) - sequence_length):
    result.append(mid_prices[index:index+sequence_length])
    print(index,':',index+sequence_length)
```
